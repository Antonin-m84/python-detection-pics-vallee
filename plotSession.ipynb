```python
import pandas as pd
import numpy as np

data = pd.read_csv("session_log.csv", delimiter=';', header=None)

data.head()
```

```python
%matplotlib qt

import matplotlib as mpl
import matplotlib.pyplot as plt

plt.figure(figsize=(100, 12))

plt.plot(data[0], data[10])
plt.plot(data[0], data[15], color="blue")
plt.plot(data[0], data[51], color="red")
plt.plot(data[0], data[56], color="yellow")

plt.ion()

plt.xlabel('0')
plt.ylabel('1')
plt.title('Plot of DataFrame')
plt.grid(True)
plt.show()
```
